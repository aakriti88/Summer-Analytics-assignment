```python
# Removing rows where horsepower is missing to reduce error
df_clean = df.dropna(subset=['horsepower'])
```

```python
#For Question 8 -  Get the row(s) with the maximum horsepower
max_hp = df_clean['horsepower'].max()
car_max_hp = df_clean[df_clean['horsepower'] == max_hp]
print("Car with highest horsepower:")
print(car_max_hp[['name', 'horsepower']])
```

```python
# For Question 9 - Number of cars with mpg ≥ 35
high_mpg = df[df['mpg'] >= 35]
print("Q9: Number of cars with mpg ≥ 35:", len(high_mpg))
```

```python
# For Question 10 - Most common origin (horsepower > 100 and weight < 3000)
filtered = df_clean[(df_clean['horsepower'] > 100) & (df_clean['weight'] < 3000)]
origin_counts = filtered['origin'].value_counts()
print("Q10: Most common origin for horsepower > 100 and weight < 3000:")
print(origin_counts)
```

```python
# For Question 11 - Mean acceleration of Japanese cars (rounded to 2 decimals)
japan_cars = df[df['origin'] == 'japan']
mean_acc = round(japan_cars['acceleration'].mean(), 2)
print("Q11: Mean acceleration of Japanese cars:", mean_acc)
```

```python
# For Question 12 - Year with highest average mpg
avg_mpg_by_year = df.groupby('model_year')['mpg'].mean()
best_year = avg_mpg_by_year.idxmax()
print("Q12: Year with highest average mpg:", best_year)
```
